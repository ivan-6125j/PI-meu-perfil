```python
top_filmes_mais_likes = df[['Film_title', 'Likes']].sort_values(by='Likes', ascending=False).head(5)
print(top_filmes_mais_likes)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=top_filmes_mais_likes, x='Likes', y='Film_title', palette='mako')
plt.title('Top 5 generos de filmes com mais Likes')
plt.xlabel('Número de Visualizações (Genres)')
plt.ylabel('Título do Filme')
plt.tight_layout()
plt.show()
```

```python
top_filmes_melhores_avaliados = df[['Film_title', 'Average_rating']].sort_values(by='Average_rating', ascending=False).head(3)
print(top_filmes_melhores_avaliados)
```

```python
top_filmes_melhores_avaliados = df[['Film_title', 'Average_rating']].sort_values(by='Average_rating', ascending=False).head(5)
print(top_filmes_melhores_avaliados)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=top_filmes_melhores_avaliados, x='Average_rating', y='Film_title', palette='mako')
plt.title('Top 5 Filmes Mais Assistidos')
plt.xlabel('Número de Visualizações (Watches)')
plt.ylabel('Título do Filme')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))
ax = sns.barplot(data=top_filmes_melhores_avaliados, x='Average_rating', y='Film_title', palette='mako')
plt.title('Top 5 Filmes Mais Bem Avaliados')
plt.xlabel('Nota Média (Average Rating)')
plt.ylabel('Título do Filme')

# Adicionando os rótulos (labels) nas barras
for i, v in enumerate(top_filmes_melhores_avaliados['Average_rating']):
    ax.text(v + 0.01, i, f'{v:.2f}', color='black', va='center')

plt.tight_layout()
plt.show()
```
